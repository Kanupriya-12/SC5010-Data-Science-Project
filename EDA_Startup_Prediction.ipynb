### Analysing Numerical Distributions

```python
f, axes = plt.subplots(6, 3, figsize=(20, 24))
count = 0

numerical_variables = ['avg_participants', 'funding_total_usd','relationships', 'age_last_milestone_year', 'age_first_milestone_year', 'milestones']

for var in numerical_variables:
    
    if startup[var].dtype != 'object':
        sb.distplot(startup[var], ax = axes[count,0])
        sb.violinplot(startup[var], ax = axes[count,1])
        sb.boxplot(startup[var], orient = "h", ax = axes[count,2])
        count += 1
print(count)
    

        
```

```python
numerical_data = startup[numerical_variables]
```

```python
numerical_data.skew()
```

### Analysing Categorical Variables

##### Most categorical variables have the dtype object. These have been filtered


```python
categorical = startup.select_dtypes('object').columns.tolist()
print(categorical)

```

```python

# REMOVING VARIABLES THAT CAN BEST BE REPRESENTED DIFFERENTLY/MAY NOT BE RELEVANT 
remove = ['zip_code','city','state_code', 'id','name', 'founded_at', 'closed_at','Unnamed: 6','state_code.1','object_id', 'first_funding_at','last_funding_at']



for label in remove:
    categorical.remove(label)
    
# APPENDING CATEGORICAL VARIABLES DISGUISED IN THE INT DTYPE (BINARY VARIABLES)
for column in startup.columns:
    if 'is_' in column or '_has' in column:
        categorical.append(column)
print(categorical)
```

```python
categorical.remove('status')
```

```python
for var in categorical:
    sb.catplot(y = var, data = startup, kind = "count", height = 8,aspect = 0.7)
```

#### Analysing categorical variables against predictor variable Status 

```python
for var in categorical:
    
    # fig, ax = plt.subplots()

    ax = sb.catplot("status", hue=var, data=startup, kind="count", legend=True)



    plt.title("Rating and Employement Bar Plot")
```
